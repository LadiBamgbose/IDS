>List of countries with highest and lowest average mortality rates: (years 2010-2015)
>>Code: WITH AvgMortalityByCountry AS (
    SELECT
        Country,
        AVG(Adult_Mortality) AS AverageMortalityRate
    FROM
        LifeExpectancy
    WHERE
        Year BETWEEN 2010 AND 2015
    GROUP BY
        Country
)

>>SELECT
    Country,
    AverageMortalityRate
FROM
    AvgMortalityByCountry
WHERE
    AverageMortalityRate = (SELECT MAX(AverageMortalityRate) FROM AvgMortalityByCountry)
    OR AverageMortalityRate = (SELECT MIN(AverageMortalityRate) FROM AvgMortalityByCountry);
